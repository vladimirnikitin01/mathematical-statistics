## Задача 1.

В этой задаче нужно визуализировать доверительные интервалы для выборки из равномерного распределения.

Чтобы не плодить код, напишите следующую функцию (см. ниже). При выборе стиля графика помните, что если изображаются лишь точки и линии, то лучше использовать серый фон, а если присутствуют закрашенные области, то предпочтительнее белый. Графики первого типа еще называют "легкими", а второго — "тяжелыми".

```python
def draw_confidence_interval(
    left, right, estimation=None, sample=None, 
    ylim=(None, None), color_estimation='#FF3300', color_interval='#00CC66', 
    color_sample='#0066FF', label_estimation='Оценка', title='Доверительный интервал',label_interval='Дов. интервал',
    label_sample='Выборка'
):
    '''
    Рисует доверительный интервал и оценку в зависимости от размера выборки.
    
    :param left: левые границы интервалов (в зависимости от n)
    :param right: правые границы интервалов (в зависимости от n)
    :param estimation: оценки (в зависимости от n)
    :param sample: выборка
    :param ylim: ограничение вертикальной оси
    :param color_estimation: цвет оценки
    :param color_interval: цвет интервала
    :param color_sample: цвет выборки
    :param label_estimation: подпись для оценки
    '''
    size=len(sample)
    n_list=[i+1 for i in range(size)]
    
    plt.figure(figsize=(18, 7.5))
    plt.grid ()
    
    estimation=np.atleast_2d(estimation)
    left=np.atleast_2d(left)
    right=np.atleast_2d(right)
    
    for j in range(estimation.shape[0] if np.any(estimation) else 0):
        plt.plot(n_list, estimation[j],
                 color=color_estimation[j] if estimation.shape[0]>1 else color_estimation,
                label=label_estimation[j] if estimation.shape[0]>1 else label_estimation)
    
    for j in range(left.shape[0]):
        plt.fill_between(n_list,left[j],right[j],
                         color=color_interval[j] if left.shape[0]>1 else color_interval,
                         alpha=0.2, label=label_interval[j] if left.shape[0]>1 else label_interval)
        
    plt.scatter(n_list, sample, label=label_sample, color=color_sample)
    
    plt.xlabel('Размер выборки', fontsize=16)
    plt.legend()
    plt.title(title, fontsize=16)
    plt.ylim(ylim)
```

Пусть $X = (X_1, ..., X_n)$ &mdash; выборка из распределения $U[0, \theta]$. Постройте асимптотические доверительные интервалы: Вальда и на основе аппроксимации $X_{(n)}$ некоторым распределением. Для этого сгенерируйте выборку $X_1, ... X_{N}, N = 100$ и постройте график доверительных интервалов уровня доверия $0.95$, вычисленных для всех подвыборок размера $n$ вида $X_1, ... X_n$, $1 \le n \le 100$, используя написанную выше функцию. Нужно нанести на график точки выборки.

Для вычисления квантилей у каждого распределения из `scipy.stats` используйте функцию `ppf`.

Для двух статистик, используемых при построении интервалов, запишите аппроксимацию некоторым распределением неформально в виде "статистика по выборке достаточно большого размера приближенно имеет такое-то распределение". 

Сделайте вывод о том, как влияет на ширину интервала вид статистики и аппроксимация.

**Решение:**

Давайте для начала напишем итоговые формулы для асимптотических доверительных интервалов. 
В интервале Вальда мы получаем $(2\overline{x}\pm z_{\frac{1+\alpha}{2}}\frac{2\overline{x}}{\sqrt{3n}})$. Теперь на основе на основе аппроксимации $X_{(n)}$ (делали это на семинаре) $(X_{(n)};X_{(n)}\frac{1}{1-\frac{u_{\alpha}}{n}})$, где $u_{\alpha}$-альфа квантиль распределения Exp(1)

```python
#Давайте возмем theta=1
theta=1
size_sample=100
n_list=np.arange(1,size_sample+1)
alpha=0.95
sample=sps.uniform(loc=0, scale=theta).rvs(size=size_sample)
estimation=2*np.cumsum(sample)/n_list

#давайте теперь вычислим интервал вальда
z=sps.norm.ppf((alpha+1)/2)
interval_valda_l=estimation-z*estimation/np.sqrt(3*n_list)
interval_valda_r=estimation+z*estimation/np.sqrt(3*n_list)

#теперь давайте на основе аппроксимации $X_{(n)}$ некоторым распределением, где u1,y2 квантили Exp(1) 0 and a уровней
u1=sps.expon.ppf(q=0,scale=1)
u2=sps.expon.ppf(q=alpha,scale=1)
x_n=np.maximum.accumulate(sample)
left_2=x_n
right_2=x_n/(1-u2/n_list)

draw_confidence_interval(left=[interval_valda_l,left_2], right=[interval_valda_r,right_2],
                         estimation=estimation,sample=sample, color_interval=['g','orange'],
                         label_interval=['Дов. интервал Вальда',r'Дов. интервал на основе аппроксимации $X_{(n)}$'],
                        title=r'Доверительный интервал $U[0,\theta]$ от размера выборки',
                        label_estimation=r'Оценка $2\overline{x}$', ylim=(0,1.75),
                        label_sample='Выборка U[0,1]')
```

**Вывод:** Давайте для начала посмотрим на то, как меняется длина полученных интервалов. В таком случае в интервале Вальда видно, что длина интервала уменьшается, как $\frac{const}{\sqrt{n}}$, в то время как длина второго интервала уменьшается как $\frac{const}{n}$. Теперь давайте запишем аппроксимацию двух статистик некоторым распределением, как просят нас в задании. В первом случае мы используем статистику $2\overline{x}$, она аппроксимируется примерно $N(\theta,\frac{\theta^2}{3n})$. А $X_{(n)}$ уже аппроксимируется $Exp(\frac{1}{\theta*n})$. Здесь тоже хорошо видно, как меняется длина интервалов. В первом случае стандартное отклонение уменьшается как 1 делить на корень из n(так меняется стандартное отклонение), во втором случае это зависимость как 1 делить на n. 

Так что во 2 случае интервал намного быстрее сходится, особенно это бросается, если мы рассмотрим ширину интервала при $n=100$. Сильно бросается в глаза, что ширина оранжевого интервала на порядок меньше, чем ширина зеленого интревала.

Также я хотел бы заметить, что при маленьких $n$ у нас второй доверительный интервал(на основе аппроксимации $X_{(n)}$) имеет большие и резкие скачки, но это логично, так как $X_{(n)}$ меняется, если в выборке появляется более большое число, таким образом может появиться большой и резкий скачок (что плохо), но начиная с $n=20$, премущества второго метода уже очевидны.

### Задача 2.
Аналогично задаче 1 сгенерируйте выборку $X_1, ... X_{100}$ из стандартного распределения Коши и постройте доверительные интервалы для следующих случаев:

* точный доверительный интервал минимальной длины в параметрической модели $\mathcal{N}(\theta, 1)$ (см. замечание ниже);
* точный асимптотический доверительный интервал в параметрической модели распределения Коши со сдвигом, используя выборочную медиану;
* точный асимптотический доверительный интервал в параметрической модели распределения Коши со сдвигом, используя асимптотически эффективную оценку.

Изобразите интервалы *на одном* графике полупрозрачными цветами. Точки выборки нужно нанести на график. 

*Замечание:*

1. Первый пример призван проиллюстрировать, что бывает, если используется неправильная модель. На практике вы никогда не знаете, из какого семейства распределений на самом деле получены данные.
**Решение:**

Для начала надо вспомнить, что такое асимптотически эффективная оценка, это одношаговая оценка выборочной медианы (выводили на лекции). Теперь давайте для каждого из 3 случаев запишем интервалы.

1)$\overline{x}\pm z_{\frac{1+\alpha}{2}}\frac{1}{\sqrt{n}}$

2)$\widehat{\mu}\pm z_{\frac{1+\alpha}{2}}\frac{\pi}{2\sqrt{n}}$ (3 номер в дз, используется асимптотический доверительный инетрвал Вальда+знания о выборочной медиане)

3) Пусть $\widehat{\theta_1}=\widehat{\mu}+\frac{\sum\limits_{i=1}^n\frac{x_i-\widehat{\mu}}{1+(x_i-\widehat{\mu})^2}}{\sum\limits_{i=1}^n\frac{1-(x_i-\widehat{\mu})^2}{(1+(x_i-\widehat{\mu})^2)^2}}$ )(получили одношаговую оценку выборочной медианы), мы знаем, что $\widehat{\theta_1}$-а.н.о. с асимптотической дисперсией 2 (теорема с лекции), тогда получаем доверительный интервал $$\widehat{\theta_1}\pm z_{\frac{1+\alpha}{2}}\frac{\sqrt{2}}{\sqrt{n}}$$

Тогда давайте все реализовывать.

```python
def one_step_eval(theta0,sample):
    #возвращает одношазовую оценку
    res=np.array([])
    x=np.array(sample)
    
    for i in range(len(sample)):
        theta_i=theta0[i]
        x_i_minus_theta=x[0:i+1]-theta_i
        numerator=np.sum(x_i_minus_theta/(1+x_i_minus_theta**2))
        denominator=np.sum((1-x_i_minus_theta**2)/((1+x_i_minus_theta**2)**2))
        res=np.append(res, numerator/denominator+theta_i)
        
    return res
```

```python
size_sample=100
sample=sps.cauchy.rvs(size=size_sample)
n_list=np.arange(1,size_sample+1)

#точный доверительный интервал минимальной длины в параметрической модели $\mathcal{N}(\theta, 1)$
estimation=np.cumsum(sample)/n_list
left_1=estimation-z*1/np.sqrt(n_list)
right_1=estimation+z*1/np.sqrt(n_list)

#точный асимптотический доверительный интервал в параметрической модели распределения Коши со сдвигом,
#используя выборочную медиану;
median=np.array([])
for i in range(size_sample):
    median=np.append(median, np.quantile(sample[:i+1], 0.5))
    
left_2=median-z*np.pi/(np.sqrt(n_list)*2)
right_2=median+z*np.pi/(np.sqrt(n_list)*2)

#точный асимптотический доверительный интервал в параметрической модели распределения Коши со сдвигом,
#используя асимптотически эффективную оценку.
best_est=one_step_eval(median,sample)
left_3=best_est-z*np.sqrt(2)/np.sqrt(n_list)
right_3=best_est+z*np.sqrt(2)/np.sqrt(n_list)
```

```python
draw_confidence_interval(left=[left_1,left_2,left_3], right=[right_1,right_2,right_3],
                         estimation=[median,best_est], label_estimation=['Выборочная медиана','Одношаговая оценка выборочной медианы'],
                         color_estimation=['r','black'],
                         sample=sample,ylim=(-5,5), color_interval=['blue','orange','g'],
                         label_interval=[r'точный доверительный интервал минимальной длины в параметрической модели $\mathcal{N}(\theta, 1)$',
                                         'точный асимптотический доверительный интервал в параметрической модели распределения Коши со сдвигом,используя выборочную медиану',
                                         'точный асимптотический доверительный интервал в параметрической модели распределения Коши со сдвигом,используя асимптотически эффективную оценку'],
                        title=r'Доверительный интервал в параметрической модели распределения Коши со сдвигом',
                        label_sample='Выборка из стандартного распределения Коши')
```

Как хорошо видно, точный доверительный интервал минимальной длины в параметрической модели $\mathcal{N}(\theta, 1)$ хорошо показывает, как важно выбирать нужную модель, так как в данном случае мы использовали неправильную модель. Теперь давайте построим более подробный график и уже без неправильной модели.

```python
draw_confidence_interval(left=[left_2,left_3], right=[right_2,right_3],
                         estimation=[median,best_est], label_estimation=['Выборочная медиана','Одношаговая оценка выборочной медианы'],
                         color_estimation=['r','black'],
                         sample=sample,ylim=(-0.4,0.8), color_interval=['orange','g'],
                         label_interval=['точный асимптотический доверительный интервал в параметрической модели распределения Коши со сдвигом,используя выборочную медиану',
                                         'точный асимптотический доверительный интервал в параметрической модели распределения Коши со сдвигом,используя асимптотически эффективную оценку'],
                        title=r'Доверительный интервал в параметрической модели распределения Коши со сдвигом',
                        label_sample='Выборка из стандартного распределения Коши')
```

2) также хорошо видно, что ширина интервала, который реализован с помощью асимптотически эффективной оценки, меньше чем ширина первого интервала. Это очевидно, так как дисперсия асимптотически эффективной оценки меньше, чем выборочной медианы, а в формулу асиптотически дов. интервала мы подставляем(теорема с лекции) стандартное отклонение нужной нам оценки (одношаговой оценки выборочной медианы или же саму выборочную медиану).

## Задача 3.
 В данном задании вам нужно изучить доверительные интервалы для параметра сдвига в нормальной модели в случае неизвестной дисперсии. Требуется построить: 
 * асимтотический доверительный интервал при помощи центральной предельной теоремы и леммы Слуцкого;
 * точный неасимптотический при помощи распределений хи-квадрат, Стьюдента.

Вывод этих интервалов был разобран на лекции. Выпишите только ответы.

Асимптотический доверительный интервал: ($\overline{x}\pm z_{\frac{1+\alpha}{2}}\cdot\frac{S}{\sqrt{n}}$)

Точный доверительный интервал: ($\overline{x}\pm T_{n-1,\frac{1+\alpha}{2}}\cdot\frac{S}{\sqrt{n-1}}$)

Реализуйте функции построения этих интервалов по выборке. Задокументируйте функции (см. [гайд](https://realpython.com/documenting-python-code/)).

```python
def calculate_asymptotic_confidence_intervals(sample, alpha=0.95):
    '''
    Строит асимтотический доверительный интервал при помощи центральной предельной теоремы и леммы Слуцкого,
    где мы получаем доверительный интервал уровня доверия alpha по выборке sample
    
    Параметры
    ----------
    sample: выборка
    alpha: уровень доверия для асимптотически доверительного интревала, по умолчанию равно 0.95

    Возвращаемое значение
    -------
    (a,b)-где a левая граница интервала, b-правая граница интервала (a,b-массивы зависимости от n)
    """
    '''
    z=sps.norm.ppf((alpha+1)/2)
    n_list=np.arange(1,len(sample)+1)
    x_mean=np.cumsum(sample)/n_list
    
    S_square=np.array([])
    for i in range (len(sample)):
        dif=np.array(sample[:i+1])-x_mean[i]
        dif=dif**2
        S_square=np.append(S_square,np.sum(dif)/(i+1))
    
    #Надо не забыть, что в формулу мы подставляем S, а не S^2
    S=np.sqrt(S_square)    
    step=S/np.sqrt(n_list)*z
    return (x_mean-step,x_mean+step)

def calculate_confidence_intervals(sample, alpha=0.95):
    '''
    Строит точный неасимптотический при помощи распределений хи-квадрат, Стьюдента,
    где мы получаем доверительный интервал уровня доверия alpha по выборке sample
    
    Параметры
    ----------
    sample: выборка
    alpha: уровень доверия для точного неасимптотического доверительного интревала, по умолчанию равно 0.95

    Возвращаемое значение
    -------
    (a,b)-где a левая граница интервала, b-правая граница интервала (a,b-массивы зависимости от n)
    """
    '''
    n_list=np.arange(1,len(sample)+1)
    x_mean=np.cumsum(sample)/n_list
    T=sps.t.ppf((alpha+1)/2,df=(n_list-1))
    
    S_square=np.array([])
    for i in range (len(sample)):
        dif=np.array(sample[:i+1])-x_mean[i]
        dif=dif**2
        S_square=np.append(S_square,np.sum(dif)/(i+1))
    
    #Надо не забыть, что в формулу мы подставляем S, а не S^2
    S=np.sqrt(S_square)
    
    warnings.filterwarnings("ignore", category=RuntimeWarning) 
    step=S/(np.sqrt(n_list-1))*(T)
    
    return (x_mean-step,x_mean+step)
```

Сгенерируйте выборку из нормального расределения и сравните два доверительных интервала в зависимости от размера выборки. Для сравнения отобразите оба интервала на одном графике. Проследите за тем, чтобы было видно, как соотносятся размеры интервалов. Поясните теоретическую причину такого поведения доверительных интервалов.

```python
#Давайте возмем theta=0, sigma=1
theta=0
sigma=1

size_sample=30
n_list=np.arange(1,size_sample+1)
sample=sps.norm(loc=theta, scale=sigma).rvs(size=size_sample)
x_mean=np.cumsum(sample)/n_list
```

```python
left_2,right_2=calculate_confidence_intervals(sample)
left_1,right_1=calculate_asymptotic_confidence_intervals(sample)

draw_confidence_interval(left=[left_1,left_2], right=[right_1,right_2],
                        sample=sample, color_interval=['g','orange'],
                        label_interval=['Асимптотический доверительный интервал',r'Точный доверительный интервал'],
                        title=r'Доверительный интервал для $\theta$ в $N(\theta,\sigma^2)$, где $\sigma$ неизвестна',
                        label_estimation=r'Оценка $\overline{x}$',estimation=x_mean)
```

Теперь давайте построим тот же самый доверительный интервал в увеличенном масштабе, чтобы посмотреть, как соотносится ширина интервалов при более больших $n$

**Вывод:** как хорошо видно, что асимптотический дов. интервал лежит внутри точного дов. интервала (особенно огромная разница при маленьких $n$). На это влияют несколько факторов.

2) Так как $n$ маленькое, а в одном случае мы делим на $\sqrt{(n)}$, а в другом на $\sqrt{n-1}$, при маленьких $n$ разница при делении все же довольно существенна. 

3) Также в точном дов. интервале мы используем распределение Стьюдента, как мы хорошо знаем при степени свободы 1-это распределение Коши, при степени свободы бесконечность-стандартное нормальное, так что с увеличением $n$ мы приближаемся к стандартному нормальному, которое используется в вычислении асимптотического интервала.

## Задача 4.
На сегодняшний день возобновляемые источники энергии становятся все более востребованными. К таким источникам относятся, например, ветрогенераторы. Однако их мощность очень трудно прогнозировать. В частности, выработка энергии при помощи ветрогенератора сильно зависит от скорости ветра. Поэтому предсказание скорости ветра является очень важной задачей. Скорость ветра часто моделируют с помощью распределения Вейбулла, которое имеет плотность:
$$p_\theta(x) = \frac{kx^{k-1}}{\lambda^k} e^{-(x/\lambda)^k} I \{x \geq 0\},$$
где $\theta = (k, \lambda)$ &mdash; двумерный параметр. К сожалению, найти точную оценку максимального правдоподобия на $\theta$ не получится.  В данном задании нужно найти оценку максимального правдоподобия приближенно с помощью поиска по сетке.

За распределение Вейбулла отвечает класс `weibull_min` из модуля `scipy.stats`, которое задается так: `weibull_min(c=`$k$ `, scale=`$\lambda$ `)`.

**Выборка:** Создайте выборку по значеням среднесуточной скорости ветра на некоторой местности для нескольких лет (не менее трех). Данные дотупны на вики. Можно выбрать любой файл.

*a).* Найдите оценку максимального правдоподобия параметра $\theta = (k, \lambda)$ с точностью $10^{-5}$ при помощи поиска по двумерной сетке.

Двумерную сетку можно создать с помощью функции `numpy.mgrid[from:to:step, from:to:step]`. Если попробовать сразу создать сетку с шагом $10^{-5},$ то может не хватить памяти. Поэтому найдите сначала максимум по сетке с большим шагом, затем сделайте сетку с маленьким шагом в окрестности найденной точки. При вычислении без циклов, возможно, придется создавать трехмерные объекты.

Функция `numpy.argmax` выдает не очень информативный индекс, поэтому пользуйтесь следующей функцией.

```python
def cool_argmax(array):
    return np.unravel_index(np.argmax(array), array.shape)
```

Нарисуйте график плотности с параметрами, соответствующим найденным ОМП, а так же нанесите на график гистограмму.

**Решение:**

```python
#давайте сначала сгенерим выборку для проверки нашего алгоритма
sample=sps.weibull_min(c=5,scale=1).rvs(size=10**3)
```

```python
x, y=np.meshgrid(np.arange(3, 8, 1e-2), np.arange(1e-3, 3, 1e-2))
```

```python
likelihood=np.sum(np.log(sps.weibull_min(c=x.reshape(-1, 1),scale=y.reshape(-1,1)).pdf(sample)), axis =1)
```

Как хорошо видно, наш поиск хорошо работает (надо шаг сделать меньше, тогда все будет работать намного точнее), теперь давайте запустим его на реальных данных

```python
data = pd.read_csv('1.csv')
data.head()
```

```python
df=data[data.Year>=2010]['Mean'].dropna().to_numpy()
df
```

```python
#теперь давайте уже сделаем поиск по сетке
step=1
start_x,end_x,start_y,end_y=1,10,1,10
accuracy=1e-6 #хотим, чтобы точность была не меньше такой
#будем увеличывать каждый раз точность на разряд, чтобы не создавать сразу огромную сетку, чтобы экономить оперативную память.
while (step>accuracy):
    x, y=np.meshgrid(np.arange(start_x, end_x, step), np.arange(start_y, end_y, step))
    likelihood=np.sum(np.log(sps.weibull_min(c=x.reshape(-1, 1),scale=y.reshape(-1,1)).pdf(df)), axis =1)
    preliminary_x=x[cool_argmax(likelihood.reshape(x.shape))]
    preliminary_y=y[cool_argmax(likelihood.reshape(y.shape))]
    
    start_x=preliminary_x-step
    end_x=preliminary_x+step
    start_y=preliminary_y-step
    end_y=preliminary_y+step
    step/=10
```

```python
print(r'Оценка максимального правдоподобия параметра theta = (k, lambda)=({},{}) с точностью 5 знаков после запятой'.format(
    '%.5f' %preliminary_x,'%.5f' %preliminary_y))
```

*b).* Обозначим $\widehat{\theta} = \left(\widehat{\lambda}, \widehat{k}\right)$ &mdash; ОМП. Запишите уравнение правдоподобия, приравняв все частные производные в точке экстремума логарифмической функции правдоподобия к $0$. Используя одно из равенств, можно выразить $\widehat{\lambda}$ через значения $X_1, \dots, X_n, \widehat{k}$; подставив это выражение в другое равенство, получить уравнение на $\widehat{k}$. Решите это уравнение приближенно с помощью метода Ньютона и получите $\widehat{k}$, а значит, и $\widehat{\lambda}$.

**Решение:**

Осталось выписать шаг в методе Ньютона для решения $f(x)=0$ $$x_{k+1}=x_k-\frac{f(x_k)}{f'(x_k)}.$$

```python
#для начала напишем f(k) и f'(k), тут у нас k переменная, а не шаг (как в формуле сверху)
def f(k,sample):
    n=len(sample)
    sample=np.array(sample)
    return 1/k+np.sum(np.log(sample))/n-(np.sum((sample**k)*np.log(sample)))/(np.sum(sample**k))

def f_derivative(k,sample):
    n=len(sample)
    sample=np.array(sample)
    return (-1/(k**2)-(np.sum((sample**k)*((np.log(sample))**2)))/(np.sum(sample**k))+
            (np.sum((sample**k)*np.log(sample)))**2/((np.sum(sample**k))**2))
```

```python
def Newton_method(sample,k=1, count_iter=int(10e5), eps_stop=1e-6):
    x_i=k
    for i in range(count_iter):
        x_i_new=x_i-f(x_i,sample)/f_derivative(x_i,sample)
        if (np.abs(x_i-x_i_new)<eps_stop):
            x_i=x_i_new
            break
            
        x_i=x_i_new
    return x_i
```

```python
k=Newton_method(df)
print(r'Оценка максимального правдоподобия параметра с помощью метода Ньютона k=({}) с точностью 5 знаков после запятой'.format(
    '%.5f' %k))
```

```python
#Как видим у нас совпали k через сетку и метод Ньютона
def search_lambda(k,sample):
    sample=np.array(sample)
    return (np.mean(sample**k))**(1/k)
```

```python
lambda_res=search_lambda(k,df)
print(r'Оценка максимального правдоподобия параметра с помощью метода Ньютона lambda=({}) с точностью 5 знаков после запятой'.format(
    '%.5f' %lambda_res))
```

```python
print(r'Оценка максимального правдоподобия параметра theta = (k, lambda)=({},{}) с точностью 5 знаков после запятой'.format(
    '%.5f' %preliminary_x,'%.5f' %preliminary_y))
print(r'Оценка максимального правдоподобия параметра theta = (k, lambda)=({},{}) с точностью 5 знаков после запятой'.format(
    '%.5f' %k,'%.5f' %lambda_res))
print()
```

**Вывод:** Мы нашли ОМП двумя способами, а именно через поиск по сетке, а также с помощью Метода Ньютона. Мы получили почти одинаковые результаты, так как мы можем выставлять нужную точность в методе ньютона и в сетке. Но в сетке есть свои сложности, связанные с тем, что мы изначально задаем диапазон в котором ищем, а также не можем сразу сгенерировать большую сетку, так как это занимает оч много памяти. Но при этом точность по сетке немного выше, так как мы выставляем нужную точность и запускаем поиск по сетке (и гарантированно находим с нужной точностью), в то время как в методе Ньютона у нас происходит множество арифметических операций(в том числе множество делений), и мы слегка теряем в качестве из-за машинной точности. 

То есть сетка позволяет нам найти с любой точностью, но тратит много памяти, в то время, как Метод Ньютона не требует выделять много памяти, но имеет маленькую неточность(ее разряд равен разряду машинной точности) из-за того, что существует такое понятие, как машинная точность.
