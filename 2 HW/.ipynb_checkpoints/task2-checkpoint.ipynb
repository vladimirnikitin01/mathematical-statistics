## Задача 1. 


**1.** Пусть $X_1, ..., X_n$ &mdash; выборка из распределения $U[0, \theta]$. Рассмотрим оценки $X_{(n)}, \frac{n+1}{n}X_{(n)}, 2\overline{X}$ параметра $\theta$.

Какие из этих оценок являются несмещенными?

Проведите практическое исследование несмещенности перечисленных оценок аналогично ислледованию несмещенности выборочной дисперсии в предыдущих заданиях. Разбор этой задачи для выборочной дисперсии был на лекции.

```python
def estimate_x_n(X):
    ''' Смещенная оценка X_(n) '''
    return X.max(axis=-1)

def estimate_x_n_corrected(X):
    ''' Несмещенная оценка (n+1)/n*X_(n)'''
    n=len(X[0])
    return np.array(X.max(axis=-1))*(n+1)/n
def estimate_2_mean(X):
    ''' Несмещенная оценка 2*X.mean()'''
    return 2*X.mean(axis=-1)
```

```python
estimators = [
    (estimate_x_n, "$X_{(n)}$", "red", 0.2),
    (estimate_x_n_corrected, "$\\frac{n+1}{n}X_{(n)}$", "green", 0.1),
    (estimate_2_mean, "$2\overline{X}$", "blue", 0.0)
]
```

```python
def est_plot(distribution, estimators, sample_size, sample_count):  
    '''
    Построение графика разброса реализаций оценок и их средних значений.
    
    distribution -- распределение формата scipy.stats
    estimators -- список оценок и параметров для отрисовки графиков
    sample_size -- размер выборок
    sample_count -- количество генерируемых выборок
    '''    

    X = distribution.rvs(size=(sample_count, sample_size))
    
    plt.figure(figsize=(18, 0.7 * len(estimators)))
    for estimator, name, color, y in estimators:
        E = estimator(X)

        plt.scatter(E, np.zeros(sample_count) + y, alpha=0.1,
                    s=100, color=color, label=name)
        plt.scatter(E.mean(), y, marker='*', s=300,
                    color='w', edgecolors='black')

    plt.vlines(1, -1, 1, color='r')
    plt.title('Размер выборки = %d' % sample_size)
    plt.yticks([])
    plt.legend()
    plt.xlim((0.6, 1.4))
    plt.ylim((-0.1, 0.1 * len(estimators)))
```

```python
sample_size_list = (10, 100, 500)  # размеры выборок
sample_count = 500  # количество экспериментов

for sample_size in sample_size_list:
    est_plot(sps.uniform, estimators, sample_size, sample_count)
```

**Вывод.** Несмещенность &mdash; свойство оценки в среднем быть равной параметру, который она оценивает. Другими словами, при бесконечном повторении эксперимента в среднем оценка будет равна истинному значению параметра.
Проведенные эксперименты подтверждают свойство несмещенности

Если посмотреть на графики, то хорошо видно, что среднее значение несмещенной оценки $\frac{n+1}{n}X_{(n)}$,$2\overline{X}$ оказались примерно равны истинному значению параметра в отличии от смещенной оценки $X_{(n)}$. Кроме того, можно заметить, что для смещенной оценки $X_{(n)}$ смещение уменьшается с ростом размера выборки, то есть она является асимптотически несмещенной. Также хорошо видно, что у $X_{(n)}$ значение близкое к истинному появляется при $n=500$, но в реальных условиях $n$ часто является маленьким числом, в таком случае нам принципиально использовать не асимптотически несмещенную оценку, а просто несмещенную оценку, если для задачи нужна несмещенность оценки.


## Задача 2.

Пусть $X_1, \ldots, X_n$ &mdash; выборка из распределения $\mathcal{N}(a, \sigma^2)$. Визуализируйте свойства состоятельности и асимптотической нормальности для оценки $S^2$ дисперсии $\sigma^2$. 

Примеры кода можно посмотреть в материалах лекций. Обязательно на графике оценок визуализируйте границы, которые задает свойство асимптотической нормальности.

**Решение:**

```python
sigma = np.sqrt(2)  # истинное значение параметра
size = 500  # размер выборок
samples_count = 500  # количество выборок
n_range = (np.arange(size) + 1)  # размеры подвыборок

# генерируем множество выборок, 
samples = np.array(sps.norm(loc=0,scale=sigma).rvs(size=(samples_count, size)))
# вычисляем оценки в зависимости от размера выборки, 
# усредняем по последней оси
estimation = (samples**2).cumsum(axis=1)/n_range-(samples.cumsum(axis=1)/n_range)**2
sqrt_var_x2=np.sqrt(2)*(sigma**2)
```

Давайте установим $\sigma^2 = 2$. Но реализация кода не фиксирует определенную $\sigma$.

```python
plt.figure(figsize=(15, 7))
for i in range(samples_count):
    plt.plot(np.arange(size) + 1, estimation[i], color='blue', alpha=0.05)
plt.plot(n_range, sigma**2 + 2*sqrt_var_x2/np.sqrt(n_range), color='red', lw=3)
plt.plot(n_range, sigma**2 - 2*sqrt_var_x2/np.sqrt(n_range), color='red', lw=3)
plt.title('Поведение оценки для разных реализаций')
plt.xlabel('Размер выборки')
plt.ylim((0, 5))
plt.xlim((0, size));
```

**Давайте приблизим и рассмотрим график при более маленьком размере выборки**

```python
plt.figure(figsize=(15, 7))
for i in range(samples_count):
    plt.plot(np.arange(size) + 1, estimation[i], color='blue', alpha=0.05)
plt.plot(n_range, sigma**2 + 2*sqrt_var_x2/np.sqrt(n_range), color='red', lw=3)
plt.plot(n_range, sigma**2 - 2*sqrt_var_x2/np.sqrt(n_range), color='red', lw=3)
plt.title('Поведение оценки для разных реализаций')
plt.xlabel('Размер выборки')
plt.ylim((0, 5))
plt.xlim((0, 50));
```

Давайте рассмотрим такую последовательность величин $Z_n = \sqrt{n}\frac{\widehat{\theta} - \theta}{\sqrt{2}\sigma^2}.$
Их распределение должно сходиться к стандартному нормальному распределению.

```python
Zn = (estimation - sigma**2) * np.sqrt(n_range) / (np.sqrt(2)*sigma**2)

plt.figure(figsize=(15, 7))
for i in range(samples_count):
    plt.plot(np.arange(size) + 1, Zn[i], color='blue', alpha=0.05)
plt.title('Траектории $Z_n$')
plt.xlabel('Размер выборки')
plt.ylim((-5, 5))
plt.xlim((0, size));
```

На графике выше можно увидеть, что при больших размерах выборки распределение похоже на стандратное нормальное. Давайте для размера выборки 500 также построим гистрограмму значений величин $Z_n$ для большей наглядности.

```python
sns.set_style('whitegrid')
grid = np.linspace(-3, 3, 100)

plt.figure(figsize=(10, 5))
plt.hist(Zn[:, -1], bins=20, density=True)
plt.plot(grid, sps.norm.pdf(grid), lw=5, label="$\\mathcal{N}(0, 1)$")
plt.legend()
plt.title('Аппроксимация $Z_n$');
```

Видно, что стандартное нормальное распределение хорошо аппроксимирует построенную гистограмму, таким образом, мы наглядно показали, что распределение $Z_n$ сходится к стандартному нормальному распределению.

**Вывод**
1. С ростом размера выборки траектории оценок сходятся к истинному значению параметра (свойство состоятельности).
2. Свойство *асимптотической нормальности* говорит о том, что с вероятностью около 0.95 при больших размерах выборки значение оценки отклоняется от истинного значения параметра не более чем на $\frac{2(\sqrt{2}\cdot\sigma^2)}{\sqrt{n}}.$ 
Эти выводы хорошо наблюдаются на графиках выше.

 Но если мы приблизим этот график в области малых размеров выборки, то тут мы можем видеть, что при малых размерах выборки асимтотические свойства не выполняются, в частности распределение оценки нельзя приблизить нормальным распределением. Приближения начинают работать начиная с размера выборки 40-50.

## Задача 3.

Пусть случайная величина $X$ имеет нормальное распределение с параметрами $(a, \sigma^2)$, то есть:
$$
X \sim \mathcal{N}(a, \sigma^2)
$$
Рассмотрим следующую задачу оптимизации:
$$
f(X) = (\mathsf{E}X-1)^2 + (\mathsf{D}X - 1)^2 \longrightarrow \min_{a, \sigma^2}
$$
В данном случае правильный ответ мы можем легко найти непосредственно, однако в реальности возникают гораздо более сложные функции, и решить задачу напрямую руками не представляется возможным. Для решения таких задач применяются различные градиентные методы, такие как, например, стохастический градиентный спуск. Оказывается, что в нём критически важно использовать несмещенные оценки градиента.

*Замечание.* Если смещение заключается только в том, что математическое ожидание оценки отличается от вектора градиента домножением каждой компоненты на одну и ту же константу, то проблем нет &mdash; все равно мы используем шаг градиента.

В данной задаче вам предлагается на примере простой функции убедиться, насколько важным оказывается использовать несмещенные оценки в итерационных процедурах.

**Решение:**

Запишите оптимальные значения параметров $a$ и $\sigma^2$, а также шаг простого градиентного спуска для минимизации определенной выше функции $f(X)$:

Реализуйте метод простого градиентного спуска для выше описанной задачи оптимизации. Для каждой итерации сохраните текущие значения среднего и дисперсии, постройте график зависимости значений $a$ и $\sigma^2$ от шага процедуры. Наблюдается ли сходимость к оптимальным параметрам?

```python
def function(x,y):
    return (x-1)**2+(y-1)**2
```

```python
# давайте начальные точки будем генерировать из равномерного распределения в области
#в которой по нашему предположению может лежать ответ
start_x,start_y=sps.uniform(1,3).rvs(size=2)
epsilon=1e-8
lambda_step=1e-4

x_array=np.array([start_x])
y_array=np.array([start_y])
while (True):
    x_new=x_array[-1]-lambda_step*(2*x_array[-1]-2)
    y_new=y_array[-1]-lambda_step*(2*y_array[-1]-2)
    
    x_array=np.append(x_array,x_new)
    y_array=np.append(y_array,y_new)
    
    if (abs(function(x_new,y_new)-function(x_array[-2],y_array[-2]))<epsilon):
        break
```

```python
fig,axs = plt.subplots(2, 1,figsize=(20,15),sharex=True)

all_steps=np.array([i for i in range(len(x_array))])

axs[0].plot(all_steps,x_array,label = "$x=a$")
axs[0].set_title("зависимости значений $a$ от шага процедуры")

axs[1].plot(all_steps,y_array,label = "$y=\sigma^2$")
axs[1].set_title("Зависимость значений $\sigma^2$ от шага процедуры")


plt.xlabel('Номер шага в градиентном спуске')
for i in range (2):
    axs[i].legend()
```

У нас очевидно видно сходимость к оптимальным параметрам, там же тут можно поэкспериментировать c epsilon, lambda_step. От этого зависит скорость и точность "схождения"

Теперь предположим, что мы хотим оценить градиент стохастически. Например, давайте текущие значения среднего и дисперсии оценивать изученным ранее методом Монте-Карло. А именно, каждый раз мы будем генерировать выборку размера 5 из нормального распределения с текущими значениями параметров и далее по ней оценивать градиент.
Для оценки математического ожидания ипользуйте несмещенную оценку, а для дисперсии &mdash; смещенную. 

Реализуйте описанный выше подход. Как и прежде, изобразите текущие значения параметров в зависимости от итерации. Сошлась ли такая процедура к оптимальным значениям?

```python
def stochastic_radient_escent(biased_estimate_var):
    x_array=np.array([start_x])
    y_array=np.array([start_y])
    
    ddof_coefficient=0
    if (biased_estimate_var==False):
        ddof_coefficient=1
    while (True):
        samples = np.array(sps.norm(loc=x_array[-1],scale=np.sqrt(y_array[-1])).rvs(size=5))
        E_samples=np.mean(samples)#несмещнная оценка матем. ожидания 
        D_samples=np.var(samples, ddof=ddof_coefficient)# оценка дисперсии, смещенность зависит от ddof_coefficient
        
        
        x_new=x_array[-1]-lambda_step*(2*E_samples-2)
        y_new=y_array[-1]-lambda_step*(2*D_samples-2)

        x_array=np.append(x_array,x_new)
        y_array=np.append(y_array,y_new)

        if (abs(function(x_new,y_new)-function(x_array[-2],y_array[-2]))<epsilon):
            break  
        
        
    fig,axs = plt.subplots(2, 1,figsize=(20,15),sharex=True)

    all_steps=np.array([i for i in range(len(x_array))])

    axs[0].plot(all_steps,x_array,label = "$x=a$")
    axs[0].set_title("зависимости значений $a$ от шага процедуры")
    axs[1].plot(all_steps,y_array,label = "$y=\sigma^2$")
    axs[1].set_title("Зависимость значений $\sigma^2$ от шага процедуры")
    axs[0].hlines(1, xmin=0, xmax=len(x_array), color='r')
    axs[1].hlines(1, xmin=0, xmax=len(x_array), color='r')
    
    if (biased_estimate_var==True):
        axs[1].hlines(5/4, xmin=0, xmax=len(x_array), color='b')

    plt.xlabel('Номер шага в стохастическом градиентном спуске')
    for i in range (2):
        axs[i].legend()
```

```python
stochastic_radient_escent(biased_estimate_var=True)
```

При использовании смещенной оценки также изобразил горизонтальную линию $y=\frac{5}{4}$, к которой должна сходиться смещенная оценка дисперсии.

Теперь изменим нашу процедуру, взяв **несмещенную оценку для дисперсии**. Обратите внимание на параметр `ddof` в функции `np.var`. Поменялся ли результат?

```python
stochastic_radient_escent(biased_estimate_var=False)
```

**Вывод**. Давайте сначала рассмотрим последние два случая, а именно, когда мы оценивали градиент стохастически. Видна огромная разница по "качеству приближения" несмещенной и смещенной оценки $\sigma^2$ к истинному значению. Поэтому мы понимаем,что предположение, что для стохастического градиетного спуска критично важна несмещенная оценка градиента - правда. Сейчас объясню почему это так. Несмещенной оценкой $\sigma^2$ является $\frac{n}{n-1}S^2$. Но мы брали выборку из 5 элементов ($n=5$), соответственно наш коэффициент $\frac{n}{n-1}=1.25$. Получается, что отношение смещенной и несмещенной оценки равно 1.25, что является большой разницей. Получается, если каждый раз мы будем оценивать $\sigma^2$ смещенной оценкой, то мы будем сдвигаться на не очень правильный коэффициент. Тогда мы не будем точно приближаться к $\sigma^2$, что хорошо видно на одном из верхних графиков.
Также хочется сравнить обычный градиентный спуск с стохастическим градиентным спуском. Во втором у нас меньше шагов выполняется, но при этом менее точно. Но точность все равно достаточно хорошая получается, при мы получаем выигрыш в скорости алгоритма (хорошая золотая середина, если у нас будут большие данные или функции, которые зависят от десяток переменных).

## Задача 4.
В заключение вам предлагается реальная практическая задача.

На основе http://ecsocman.hse.ru/data/819/759/1219/Monte_Karlo_dlya_analitikov.pdf
 	
Рассмотрим проект по разработке нефтяного месторождения. В основе модели проекта лежат предварительные данные о величине резервов месторождения.

*Замечание.* Знания физики или экономики не требуются.



*Формулировка задачи от заказчика:*
Общая задача анализа &mdash; основываясь на величине запасов и проценте нефтеотдачи рассчитать NPV (чистая приведенная стоимость) проекта, а точнее, $0.1$-квантиль ее распределения.
Следующим этапом мы хотим использовать ее в качестве критерия оптимизации, то есть максимизировать такое значение NPV, которого мы можем достигнуть или превысить с $90\%$-й вероятностью, подобрав при этом оптимальное количество скважин на месторождении.



Предположим, что на месторождении есть $n$ скважин. Эти скважины добывают некоторую смесь, которая состоит из воды, нефти и различных примесей. Доля нефти из добытого материала называется <a href="https://ru.wikipedia.org/wiki/Нефтеотдача">коэффициентом нефтеотдачи</a> $k$. Мы будем считать, что этот коэффициент является одинаковым для всего месторождения и имеет нормальное распределение со средним $42\%$ и стандартным отклонением $1.2\%$.




**Добыча нефти скважиной за год**

<a href="http://vseonefti.ru/upstream/stadii-razrabotki.html">Разработка месторождения</a> включает три этапа:
1. фаза роста добычи &mdash; период введения в работу новых скважин;
2. фаза плато: после достижения определенного уровня добычи, она некоторое время продолжается на постоянном уровне;
3. фаза снижения добычи &mdash; период, когда темпы добычи экспоненциально снижаются с течением времени.


Для упрощения задачи мы пропустим два первых этапа и рассмотрим только последний.


Каждая скважина $j$ характеризуется параметром $q_j$ &mdash; темп добычи из скважины, определяемый объемом вещества (смесь нефти, воды и др.), добываемого скважиной за сутки. Будем считать, что этот параметр является одинаковым для скважины в течении всего периода разработки и имеет нормальное распределение со средним $10$ тыс. баррелей и стандартным отклонением $3$ тыс. баррелей. Темпы добычи для разных скважин считаются независимыми случайными величинами.

Соответственно, за год $t$ скважина добывает $Q_{tj} = 365 \cdot k \cdot q_j \cdot e^{-0.008 \cdot n \cdot (t-1)}$ тыс. баррелей нефти, где экспонента отвечает за снижение добычи с течением времени.
Всего за год $t$ на месторождении добывается $Q_t = \sum\limits_{j=1}^{n} Q_{tj}$ тыс. баррелей нефти.




**Прибыль**

Стоимость барреля нефти будем считать постоянной и равной $c = 70$ долларов за баррель. Однако, для расчета стоимости нужно учесть <a href="https://ru.wikipedia.org/wiki/Ставка_дисконтирования">ставку дисконтирования</a> &mdash; процентная ставка, используемая для пересчета будущих потоков доходов в единую величину текущей стоимости (см. формулу далее). Обозначим ее $i$ и будем считать, что она имеет нормальное распределение со средним $10\%$ и стандартным отклонением $1.2\%$.

Стоимость добытой нефти за год $t$ составит (тыс. баррелей)
$$\frac{c \cdot Q_t}{(1+i)^{t-1}}.$$

Будем считать, что разработка месторождения прекращается, если за год на всем месторождении было добыто менее $100$ тыс. баррелей нефти. Последний год разработки обозначим $T$.




**Затраты**

Затраты на месторождение (кроме скважин) составляют $C_{\text{мест.}} = 200$ млн. долларов в год. Будем считать, что издержки на содержание скважины $j$ за весь период разработки имеют треугольное распределение (<a href="https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.triang.html">scipy.stats.triang</a>) c минимальным значением $90$ млн. долларов, максимальным значением $120$ млн. долларов и модой (наиболее вероятное значение) $100$ млн. долларов. Обозначим эти случайные величины $C^j_{\text{скв.}}$ и будем считать их независимыми.




**NPV**

Теперь мы можем написать формулу NPV
$$NPV = \sum\limits_{t=1}^T \frac{c \cdot Q_t}{(1+i)^{t-1}} - T \cdot C_{\text{мест.}} - \sum\limits_{j=1}^{n} C^j_{\text{скв.}}.$$




**Задание**

С помощью метода Монте-Карло требуется найти число $x$, при котором $$\mathsf{P}(NPV < x) = 0.1.$$

Количество итераций метода должно быть не менее $100\ 000$. На основе проделанных итераций оцените также среднее значение NPV и вероятность, с которой NPV будет положительна. Кроме того, постройте нормированную гистограмму значений NPV с помощью `plt.hist(values, bins=200, density=True)`.



Перечислим еще раз все *случайные величины*:
* Коэффициент нефтеотдачи $k$ имеет нормальное распределение со средним $42\%$ и стандартным отклонением $1.2\%$;
* $q_1, ..., q_{n}$ &mdash; темпы добычи из скважин &mdash; независимые нормальные случайные величины со средним $10$ тыс. баррелей и стандартным отклонением $3$ тыс. баррелей;
* Ставка дисконтирования $i$ имеет нормальное распределение со средним $10\%$ и стандартным отклонением $1.2\%$;
* $C^1_{\text{скв.}}, ..., C^{n}_{\text{скв.}}$ &mdash; затраты на каждую скважину &mdash; независимые  случайные величины, имеющие треугольное распределение c минимальным значением $90$ млн. долларов, максимальным значением $120$ млн. долларов и модой (наиболее вероятное значение) $100$ млн. долларов (используйте `sps.triang(loc=90, c=1/3, scale=30)`).

Параметры считать независимыми.

*Фиксированные параметры*:
* $n = 25$ скважин;
* $365$ дней в году;
* $c=70$ долларов за баррель &mdash; стоимость нефти;
* $100$ тыс. баррелей &mdash; объем добытой нефти за год, при котором разработка месторождения прекращается;
* $200$ млн. долларов в год &mdash; затраты на месторождение.

*Замечания по решению*:
* Все случайные величины надо сгенерировать заранее, до вычисления значений $T$.
* Не стоит выводить явную формулу для $T$, используйте цикл по годам.
* Запрещается использовать цикл по количеству элементов в выборке.

**Решение:**

```python
#Давайте сначала сгенерим все наши данным и введем все важные константы
#но давайте все цены поделим на миллион, чтобы затем на всех графиках все было более наглядным
n=25
days=365
c=70/1e6
limiti_Q=1e5
C_field=200
```

```python
def cost_extracted_oil(i,t,Q_t):
    return c*Q_t/((1+i)**(t-1))
def NPV(size):
    # давайте прогенерим часть нужных с.в.
    #но давайте все цены поделим на миллион, чтобы затем на всех графиках все было более наглядным
    k=(sps.norm(loc=42/100,scale=1.2/100).rvs(size=size))
    q_j=np.array(sps.norm(loc=10e3,scale=3e3).rvs(size=(size,n)))
    i=(sps.norm(loc=10/100,scale=1.2/100).rvs(size=size))
    C_j=np.array(sps.triang(loc=90, c=1/3, scale=30).rvs(size=(size,n)))
    
    #теперь пора начать вычислять первое слагаемое в формуле
    t=0
    Q_t_j=days*k*np.sum(q_j, axis=1)
    Q_sum=np.zeros(size)
    
    #теперь считаем, если хоть в одной сгенерированной выборке в месторождении есть добыча нефти
    while (Q_t_j>=limiti_Q).any():
        t+=1
        Q_t_j=days*k*np.sum(q_j, axis=1)*np.exp(-0.008*n*(t-1))
        Q_t_j[Q_t_j < limiti_Q] = 0
        
        Q_sum+=cost_extracted_oil(i,t,Q_t_j)
        #там где еще добыча идет, вычитаем стоимость обслужтвания каждого года
        Q_sum[Q_t_j >= limiti_Q] -= C_field
    
    #теперь 3 слагаемое
    Q_sum-=np.sum(C_j,axis=1)
    
    return Q_sum
```

```python
count=int(1e6) #количество итераций
all_NPV=NPV(count)
```

```python
print("Среднее значение NPV равно {} миллионов долларов".format('%.1f' %np.mean(all_NPV)))
```

```python
print(f"С вероятность {'%.2f' %(len(all_NPV[all_NPV>0])/len(all_NPV)*100)} NPV будет положительна".format())
```

```python
print(f"0.1-квантиль распределения NPV равен {'%.1f' %np.quantile(all_NPV,0.1)} миллиона")
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(20, 11))
plt.hist(all_NPV, bins=200, density=True)
plt.xlabel("$NPV$, млн. долларов")
plt.ylabel("Плотность")

#теперь осталось квантиль-0.1 отметить на графике
plt.axvline(x=np.percentile(all_NPV, 10), color='red', 
            label='0.1-квантиль = {} млн. долларов'.format('%.1f' %np.percentile(all_NPV, 10)))
plt.legend()
plt.title("Нормированная гистограмма значений NPV");
```

```python
#давайте посчитаем параметры полученного распределения. 
D=all_NPV.var(ddof=1)#используем несмещенную выборочную дисперсию
E=all_NPV.mean()#используем для несмещенной оценки X.mean()
sigma_npv=np.sqrt(D)
#хочу посмотреть на закон 2 сигм
print("С вероятностью 95% выручка проекта будет лежать в диапазоне {}÷{} миллионов.".format('%.1f' %(E-2*sigma_npv), '%.1f' %(E+2*sigma_npv)))
```

**Вывод:** Для начала мы понимаем, что нефтяной бизнес выгоден, как проект. Так как с вероятностью равной 99.76 он будет прибыльным. Также с вероятностью 90 процентов наша выруча будет больше 943.4 миллиона. Но что более важно, так то, что с вероятностью 95% выручка проекта будет лежать в диапазоне 446.0÷3319.5 миллионов. Что уже звучит как очень выгодный проект с маленькими рисками.
